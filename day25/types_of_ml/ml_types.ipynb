# Part A — Concept Application

## Problem Identification

### Classify given problems :

#### Supervised / Unsupervised / Reinforcement Learning
-	A system predicts whether an email is spam or not using past labeled emails.
-	A retail company groups customers based on purchasing behavior without any predefined labels.
-	A robot learns to walk by trying different movements and receiving rewards when it moves correctly.
-	A model predicts house prices based on features like area, location, and number of rooms.
-	An e-commerce platform recommends similar products based on user browsing history.

### Answers:

### A system predicts whether an email is spam or not using past labeled emails.
Supervised Learning

- Reason: Uses labeled data (emails labeled as spam or not) to train a model.

### A retail company groups customers based on purchasing behavior without any predefined labels.
Unsupervised Learning

- Reason: No labels are provided; the goal is to find patterns or clusters in the data.

### A robot learns to walk by trying different movements and receiving rewards when it moves correctly

Reinforcement Learning

- Reason: Learning occurs via trial and error with feedback in the form of rewards.

### A model predicts house prices based on features like area, location, and number of rooms  
Supervised Learning
- Reason: Predicting a continuous target (house price) using labeled examples.

### An e-commerce platform recommends similar products based on user browsing history.
Unsupervised Learning

- Reason: Often uses collaborative filtering or clustering, which is unsupervised, though sometimes hybrid methods may use supervised signals.

## Regression vs Classification

### Regression: 
- Predicts a continuous numeric target, such as house prices, temperature, or salary. 

Justification: 
- The target can take any numeric value within a range, so the model must estimate a number rather than a category.

### Classification: 
- Predicts a categorical target, such as spam/not spam, disease/no disease, or product category.
 
 Justification: 
- The target consists of discrete classes, so the model assigns each input to one of the predefined categories.

# Part B - Stretch Problem

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
```

## Regression

### Create a small regression dataset

```python
regression_data = {
    'Hours_Studied': [1, 2, 3, 4, 5],
    'Sleep_Hours': [7, 6, 5, 4, 3],
    'Score': [50, 55, 65, 70, 80]  # Target variable
}
df = pd.DataFrame(regression_data)
```

### Inputs (features) and output (target)

```python
X_reg = df[['Hours_Studied', 'Sleep_Hours']]
y_reg = df['Score']
```

### Split data

```python
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
```

### Train Linear Regression model

```python
reg_model = LinearRegression()
reg_model.fit(X_train_reg, y_train_reg)
```

### Predict

```python
y_pred_reg = reg_model.predict(X_test_reg)
```

```python
print("Regression Example:")
print("Input (features):\n", X_test_reg)
print("Output (predicted Score):\n", y_pred_reg)
print("Use case: Predicting student exam scores based on study and sleep hours.\n")
```

# Classification

### Create a small classification dataset

```python

classification_data = {
  'Temperature': [30, 20, 35, 15, 28, 22, 33, 18, 27],
    'Humidity': [70,  50, 80, 40, 65, 55, 75, 45, 68],
    'Play': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes']  # Target variable
}
df = pd.DataFrame(classification_data)
```

### Inputs (features) and output (target)

```python
X_clf = df[['Temperature', 'Humidity']]
y_clf = df['Play']
```

### Split data

```python
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)
```

### Train Decision Tree Classifier

```python
clf_model = DecisionTreeClassifier()
clf_model.fit(X_train_clf, y_train_clf)
```

### Predict

```python
y_pred_clf = clf_model.predict(X_test_clf)
```

```python
print("Classification Example:")
print("Input (features):\n", X_test_clf)
print("Output (predicted Play):\n", y_pred_clf)
print("Use case: Deciding whether to play outside based on temperature and humidity.")
```
